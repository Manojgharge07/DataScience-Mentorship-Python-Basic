# **Select two rows and three columns**

In order to select two rows and three columns, we select a two rows which we want to select and three columns and put it in a separate list like this:

Dataframe.loc[["row1", "row2"], ["column1", "column2", "column3"]]

```python
import pandas as pd


# retrieving two rows and three columns by loc method
first = df.loc[["Avery Bradley", "R.J. Hunter"],
				["Team", "Number", "Age", "Position","Weight"]]



print(first)
```

# **Select all of the rows and some columns**

In order to select all of the rows and some columns, we use single colon [:] to select all of rows and list of some columns which we want to select like this:

Dataframe.loc[:, ["column1", "column2", "column3"]]

```python
# retrieving all rows and some columns by loc method
import pandas as pd


data = pd.read_csv('/content/drive/MyDrive/Nba.csv/nba.csv')

first = df.loc[:, ["Team", "Number", "Age", "Salary", "Name"]]



print(first)
```

**IX FUNCTION (df.ix())**

1. Pandas DataFrame.ix[ ] is both Label and Integer based slicing technique.

2. Besides pure label based and integer based, Pandas provides a hybrid method for selections and subsetting the object using the ix[] operator.

```python
# importing pandas 
import pandas as pd
    
# making data frame from csv file
data = pd.read_csv("/content/drive/MyDrive/Nba.csv/nba.csv")  
    
```

```python
data
```

```python
# Integer slicing
print("Slicing only rows(till index 3):")
df = data.ix[:3, ]
print(df, "\n")
   
print("Slicing rows and columns(rows=3, col 1-3, excluding 3):")
df1 = data.ix[:3, 1:3]
print(df1)
```
